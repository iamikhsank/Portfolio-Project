## **Cleaning and Standardizing Data**
The data was standardized into a uniform format, ensuring consistency across each value.
and Input errors were corrected in the sales_method column to maintain data integrity, consolidating variations of identical values to be consistent.

Standardization was performed to align the data format in the sales_method column and to cap extreme values in the years_as_customer column. The following steps were applied:

Sales Method Standardization:

The symbol + was replaced with & to ensure consistency in the formatting.
Variants like em + call were standardized to Email + Call.
All entries labeled email were reformatted to capitalize the first letter, becoming Email.
The variant em & call was adjusted to Email & Call.
This step ensured that each entry in the sales_method column was formatted consistently, which facilitates easier analysis and segmentation.

Years as Customer Capping:

A maximum cap was set for the years_as_customer column at 39 years. Any values exceeding this limit were set to 39. This measure was taken to eliminate outliers that could skew analysis results.

```python
# data cleaning and Standardizing
def standardizing(df):
  # sales method
  df['sales_method'] = df['sales_method'].str.replace('+','&')
  df['sales_method'] = df['sales_method'].str.replace('em + call','Email + Call')
  df['sales_method'] = df['sales_method'].str.replace('email','Email')
  df['sales_method'] = df['sales_method'].str.replace('em & call','Email & Call')
  df.loc[df['years_as_customer'] > 39,'years_as_customer'] = 39
  return df
```

```python
df['sales_method'] = df['sales_method'].str.replace('em & call','Email & Call')
```

```python
df = standardizing(df)
```

## Missing Value Imputation
In the revenue column, some values were missing. These missing values were filled using the average revenue based on each sales_method category. Imputing missing values using the average from each category helps maintain data integrity and reduce potential bias caused by missing entries.

```python
# find missing values
df.isna().sum()
```

```python
# find revenue mean per sales method before we filled it with mean per method
df.groupby('sales_method')['revenue'].mean()
```

```python
# filling data
def fill(df):
  df['revenue'] = df['revenue'].fillna(df.groupby('sales_method')['revenue'].transform('mean').round(2))
  return df
```

## Defining Business Metrics to Track Performance - Average Revenue per Customer Sales Effort

To get a better understanding of the performance of each sales method I create a metric called Average Reveue per Customer Sales Effort(ARPSE). This metric takes into account the difficulty of each sales method in terms of time spent with email having a value of 0.5, email + call 1 and call 3.

```python
# Define the sales effort for each sales method
sales_effort = {
    'Email': 0.5,
    'Call': 3,
    'Email & Call': 1}

grouped_df = clean_df.groupby(['sales_method']).agg({'revenue':'sum','customer_id':'count'}).reset_index()

# Calculate ARCPSE for each sales method
grouped_df['ARPSE'] = grouped_df.apply(lambda row: row['revenue'] / (row['customer_id'] * sales_effort[row['sales_method']]), axis=1)
```

```python
grouped_df[['sales_method','ARPSE']].style.background_gradient(cmap='crest')
```

The ARCPSE results show that Email has the highest value with 194.25, followed by Email & Call with 183.65 and finally call with 15.86. This suggests that the call method is the most efficient in terms of reveneue generated per unit of effort.

However, it is important to note that additional factors have been observed. The 'Email + Call' method had customers with more website visits and higher average number of items purchased. Moreover, this method demonstrated a steady increase in revenue growth over the six-week period. In contrast, the 'Call' method showed high revenue in week 1 but decreased over time.
